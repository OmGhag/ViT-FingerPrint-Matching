```python
import numpy as np
import pandas as pd
import torch
from transformers import pipeline
```

```python
import cv2 as cv
```

```python
img_real = r"H:\My Drive\Biometric Dataset\archive (1)\SOCOFing\Real\1__M_Left_index_finger.BMP"
img_altered = r"H:\My Drive\Biometric Dataset\archive (1)\SOCOFing\Altered\Altered-Hard\1__M_Left_index_finger_Obl.BMP"
```

```python
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
pipe = pipeline(task="image-feature-extraction", model_name="google/vit-base-patch16-384", device=DEVICE, pool=True)
```

```python
outputs = pipe([img_real, img_altered])
```

```python
print(len(outputs[0][0]))
print(outputs)
```

```python
from torch.nn.functional import cosine_similarity

similarity_score = cosine_similarity(torch.Tensor(outputs[0]),torch.Tensor(outputs[1]), dim=1)

print(similarity_score)

```
